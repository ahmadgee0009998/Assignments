```python
# Student Data Analyzer System

# Function to add a student
def add_student():
    name = input("Enter student name: ")
    age = int(input("Enter age: "))
    marks = float(input("Enter marks: "))
    
    # Determine pass/fail using conditional operator
    status = "Pass" if marks >= 50 else "Fail"
    
    student = {
        "name": name,
        "age": age,
        "marks": marks,
        "status": status
    }
    
    return student

# Function to display all students
def display_students(students):
    print("\n--- Student Records ---")
    for i, student in enumerate(students):
        print(f"\nStudent {i+1}:")
        for key, value in student.items():
            print(f"{key}: {value}")

# Function to calculate average marks
def calculate_average(students):
    total = 0
    for student in students:
        total += student["marks"]   # arithmetic operator
    
    avg = total / len(students) if students else 0
    return avg

# Function to find topper
def find_topper(students):
    if not students:
        return None
    
    topper = students[0]
    for student in students:
        if student["marks"] > topper["marks"]:  # comparison operator
            topper = student
    
    return topper

# Function to filter passed students
def passed_students(students):
    passed = []
    for student in students:
        if student["marks"] >= 50 and student["status"] == "Pass":  # logical operator
            passed.append(student)
    return passed

# Main program
def main():
    students = []   # list data type
    
    while True:
        print("\n===== MENU =====")
        print("1. Add Student")
        print("2. Display Students")
        print("3. Calculate Average Marks")
        print("4. Find Topper")
        print("5. Show Passed Students")
        print("6. Exit")
        
        choice = input("Enter your choice: ")
        
        # Membership operator example
        if choice not in ["1", "2", "3", "4", "5", "6"]:
            print("Invalid choice! Try again.")
            continue
        
        if choice == "1":
            student = add_student()
            students.append(student)
        
        elif choice == "2":
            display_students(students)
        
        elif choice == "3":
            avg = calculate_average(students)
            print(f"Average Marks: {avg}")
        
        elif choice == "4":
            topper = find_topper(students)
            if topper:
                print("\nTopper Details:")
                print(topper)
        
        elif choice == "5":
            passed = passed_students(students)
            print("\nPassed Students:")
            for student in passed:
                print(student)
        
        elif choice == "6":
            print("Exiting program...")
            break

# Run the program
main()
```
