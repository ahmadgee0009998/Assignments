```python
#Build a simple class system using OOP.

# Instructions:
# 1. Create a base class Person with attributes name and age.
# 2. Create method display_info() that prints name and age.
# 3. Create derived class Student that inherits Person.
# 4. Add attributes student_id and major.
# 5. Override display_info() to print student details.
# 6. Create derived class GraduateStudent inheriting Student.
# 7. Add thesis_topic attribute and override display_info().
# 8. Create one object of each class and call display_info().
```

```python
#Base class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print("--- Person ---")
        print("Name:", self.name)
        print("Age:", self.age)

#Derived class: Student
class Student(Person):
    def __init__(self, name, age, student_id, major):
        super().__init__(name, age)
        self.student_id = student_id
        self.major = major

    def display_info(self):
        print("--- Student ---")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Student ID:", self.student_id)
        print("Major:", self.major)


#Derived class: GraduateStudent
class GraduateStudent(Student):
    def __init__(self, name, age, student_id, major, thesis_topic):
        super().__init__(name, age, student_id, major)
        self.thesis_topic = thesis_topic

    def display_info(self):
        print("--- Graduate Student ---")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Student ID:", self.student_id)
        print("Major:", self.major)
        print("Thesis Topic:", self.thesis_topic)

#Creating objects
person = Person("John Doe", 45)
student = Student("Jane Smith", 20, "S12345", "Computer Science")
grad_student = GraduateStudent("Bob Johnson", 27, "G67890", "Physics", "Quantum Computing")

#Calling methods
person.display_info()
student.display_info()
grad_student.display_info()
```
