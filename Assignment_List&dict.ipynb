```python
#Build a program that manages a shopping list using lists and dictionaries.

# Instructions:
# 1. Create a list named shopping_list containing at least 5 initial items (e.g., "milk", "bread", "eggs", "apples", "chicken").
# 2. Print the entire list, and also print the third item in the list.
# 3. Add two new items to the list using an appropriate list method.
# 4. Remove one item from the list.
# 5. Create a dictionary named item_prices. The keys should be items from your shopping list (at least 4), and the values should be their prices (as floats).
# 6. Print the price of one specific item from the dictionary.
# 7. Print a formatted summary of your final shopping list and the price dictionary.
```

```python
#Create a shopping list with 5 initial items
shopping_list = ["milk", "bread", "eggs", "apples", "chicken"]
# Print the entire list and the third item
print("Initial Shopping List:", shopping_list)
print("The third item is:", shopping_list[2])
#Add two new items to the list
print("Adding 'cheese' and 'coffee'...")
shopping_list.append("cheese")
shopping_list.append("coffee")
#Remove one item from the list
print("Removing 'chicken'...")
shopping_list.remove("chicken")
#updated list
print("Updated Shopping List:", shopping_list)
#Create a dictionary with item prices
item_prices = {
    "milk": 2.5,
    "bread": 1.99,
    "eggs": 3.25,
    "apples": 0.99
}
#Print the price of a specific item
print("The price of bread is $", item_prices["bread"])
#Print final result
print("\n--- Final Summary ---")
print("Items to buy:", ", ".join(shopping_list))
print("Price lookup:", item_prices)
```
