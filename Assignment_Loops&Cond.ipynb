```python
#Write a script that processes a range of numbers using loops and conditional statements.

# Instructions:
# 1. Use a for loop to iterate through numbers from 1 to 20.
# 2. For each number:
#    - If divisible by 3, print "Fizz".
#    - If divisible by 5, print "Buzz".
#    - If divisible by both 3 and 5, print "FizzBuzz".
#    - Otherwise, print the number itself.
# 3. After the loop, ask the user for a number.
# 4. Use a while loop to print a countdown from that number down to 1, and then print "Blast Off!".
# 5. Use break to stop the countdown if the user's number is greater than 10.
```

```python
#Use a for loop to iterate from 1 to 20
for num in range(1, 21):
    # Conditions
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")   # Divisible by both 3 and 5
    elif num % 3 == 0:
        print("Fizz")       # Divisible by 3
    elif num % 5 == 0:
        print("Buzz")       # Divisible by 5
    else:
        print(num)          # Otherwise print the number

#Ask the user for a number
user_num = int(input("\nEnter a number for countdown: "))
#Stop the countdown if the number is greater than 10
if user_num > 10:
    print("Number is greater than 10. Countdown stopped!")
else:
    # Use a while loop to count down
    while user_num > 0:
        print("Countdown:",user_num)
        user_num -= 1

    print("Blast Off!")
```
