```python
#Create a simple program that acts as a basic personal assistant by asking for user input and performing simple operations.

# Instructions:
# 1. Ask the user for their name and store it in a variable.
# 2. Ask the user for their birth year (as an integer) and store it.
# 3. Ask the user for their current age (as an integer) and store it.
# 4. Calculate and print the year they will turn 100 years old. (Hint: 100 - age + current_year).
# 5. Ask the user for a lucky number (as a float).
# 6. Print a summary message that includes all the information you've collected. Use appropriate data types and typecasting.
# 7. Add comments to your code explaining each step.
```

```python
#Entering user name
name=input("What is your name?:") 
#Entering user D.O.B
birth_year=int(input("What is your D.O.B?:"))
#Entering user age
age=int(input("What is your age?:"))
#Get the current year from the system 
from datetime import datetime #Get the current year from the system 
current_year = datetime.now().year 
# Calculate the year when the user will turn 100
year_turn_100 = (100 - age) + current_year
#Entering user lucky number
lucky_number = float(input("Enter your lucky number: "))
#Output
print("\n--- Personal Assistant Summary ---")
print("Hello,", name + "!")
print("You were born in", birth_year)
print("You are currently", age, "years old.")
print("You will turn 100 in the year", year_turn_100)
print("Your lucky number is", lucky_number)
```
