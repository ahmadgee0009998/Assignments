```python
# Python Fundamentals Project
# Student Management & Data Processing System

# Function to add student
def add_student(students):
    name = input("Enter name: ")
    age = int(input("Enter age: "))
    marks = float(input("Enter marks: "))
    
    # Using dictionary
    student = {
        "name": name,
        "age": age,
        "marks": marks
    }
    
    students.append(student)
    print("Student added successfully!")

# Function to display students
def display_students(students):
    if not students:
        print("No records found!")
        return
    
    print("\n--- Student List ---")
    for i, student in enumerate(students):
        print(f"\nStudent {i+1}")
        for key, value in student.items():
            print(f"{key}: {value}")

# Function to calculate average marks
def calculate_average(students):
    if len(students) == 0:
        return 0
    
    total = 0
    for s in students:
        total += s["marks"]   # arithmetic operator
    
    return total / len(students)

# Function to find highest marks
def highest_marks(students):
    if not students:
        return None
    
    top = students[0]
    for s in students:
        if s["marks"] > top["marks"]:   # comparison operator
            top = s
    return top

# Function to categorize students
def categorize_students(students):
    categories = {
        "Pass": [],
        "Fail": []
    }
    
    for s in students:
        if s["marks"] >= 50:   # conditional statement
            categories["Pass"].append(s["name"])
        else:
            categories["Fail"].append(s["name"])
    
    return categories

# Function demonstrating tuple and set
def extra_data_structures():
    subjects = ("Math", "Physics", "CS")   # tuple
    unique_ids = {101, 102, 103, 101}     # set
    
    print("\nSubjects (Tuple):", subjects)
    print("Unique IDs (Set):", unique_ids)

# Main program
def main():
    students = []   # list
    
    while True:
        print("\n===== MENU =====")
        print("1. Add Student")
        print("2. Display Students")
        print("3. Average Marks")
        print("4. Highest Marks")
        print("5. Categorize Students")
        print("6. Show Tuple & Set Example")
        print("7. Exit")
        
        choice = input("Enter choice: ")
        
        if choice == "1":
            add_student(students)
        
        elif choice == "2":
            display_students(students)
        
        elif choice == "3":
            avg = calculate_average(students)
            print("Average Marks:", avg)
        
        elif choice == "4":
            top = highest_marks(students)
            if top:
                print("\nTop Student:", top)
        
        elif choice == "5":
            result = categorize_students(students)
            print("\nPass Students:", result["Pass"])
            print("Fail Students:", result["Fail"])
        
        elif choice == "6":
            extra_data_structures()
        
        elif choice == "7":
            print("Exiting program...")
            break
        
        else:
            print("Invalid choice!")

# Run the program
main()
```
