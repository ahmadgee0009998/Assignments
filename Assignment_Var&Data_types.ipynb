```python
# Variables (Name, Age, City)

name = "Ahmad"
age = 21
city = "Gujranwala"
```

```python
# Data Types (int, float, string, boolean)

num_int = 10
num_float = 8.6
text = "Hello"
is_student = True
```

```python
# Check Data Types

print(type(name))
print(type(age))
print(type(city))
print(type(num_int))
print(type(num_float))
print(type(text))
print(type(is_student))
```

```python
# Arithmetic Operations

a = 10
b = 5

print("Addition=", a + b)
print("Subtraction=", a - b)
print("Multiplication=", a * b)
print("Division=", a / b)
```

```python
# Comparison Operators

x = 10
y = 20

print("x > y:", x > y)
print("x < y:", x < y)
print("x == y:", x == y)
```

```python
# Type Conversion

num1 = 10
num2 = 5.8

num1_float = float(num1)  # int to float

num2_int = int(num2)   # float to int

print(num1_float)
print(num2_int)
```

```python
# Student Information Program

student_name = "Ahmad"
roll_number = 101
marks = 95

print("Name:", student_name)
print("Roll Number:", roll_number)
print("Marks:", marks)
```

```python
# Assignment Operators

num = 10

num += 5   #Add 5 in 10
print(num)

num -= 3  #Subtract 3 in 15
print(num)

num *= 2   #Multiply 2 in 12
print(num)
```

```python
# Boolean Operations

a = True
b = False

print("AND Operator:", a and b)
print("OR Operator:", a or b)
print("NOT Operator:", not a)
```

```python
# Area of Rectangle

length = 5
width = 3

area = length * width
print("Area of Rectangle:", area)
```
