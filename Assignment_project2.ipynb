```python
# Smart Calculator & Number Analyzer

# Function for addition
def add(a, b):
    return a + b

# Function for subtraction
def subtract(a, b):
    return a - b

# Function for multiplication
def multiply(a, b):
    return a * b

# Function for division
def divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b

# Function to check even or odd
def check_even_odd(num):
    return "Even" if num % 2 == 0 else "Odd"

# Function to find factorial using loop
def factorial(n):
    if n < 0:
        return "Invalid input"
    result = 1
    for i in range(1, n + 1):   # for loop
        result *= i
    return result

# Function to generate multiplication table
def multiplication_table(num):
    print(f"\nMultiplication Table of {num}:")
    for i in range(1, 11):
        print(f"{num} x {i} = {num * i}")

# Function to check prime number
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num):   # loop
        if num % i == 0:
            return False
    return True

# Main program using while loop
def main():
    while True:
        print("\n===== MENU =====")
        print("1. Calculator")
        print("2. Even/Odd Checker")
        print("3. Factorial")
        print("4. Multiplication Table")
        print("5. Prime Checker")
        print("6. Exit")
        
        choice = input("Enter your choice: ")
        
        if choice == "1":
            a = float(input("Enter first number: "))
            b = float(input("Enter second number: "))
            
            print("Add:", add(a, b))
            print("Subtract:", subtract(a, b))
            print("Multiply:", multiply(a, b))
            print("Divide:", divide(a, b))
        
        elif choice == "2":
            num = int(input("Enter a number: "))
            print("Result:", check_even_odd(num))
        
        elif choice == "3":
            num = int(input("Enter a number: "))
            print("Factorial:", factorial(num))
        
        elif choice == "4":
            num = int(input("Enter a number: "))
            multiplication_table(num)
        
        elif choice == "5":
            num = int(input("Enter a number: "))
            if is_prime(num):
                print("It is a Prime Number")
            else:
                print("Not a Prime Number")
        
        elif choice == "6":
            print("Exiting program...")
            break
        
        else:
            print("Invalid choice! Try again.")

# Run the program
main()
```
